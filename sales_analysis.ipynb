```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Simulate sales and profit values
sales = np.random.randint(8000, 20000, size=12)
profit = sales * np.random.uniform(0.2, 0.4, size=12)  # assume 20%-40% profit margin

# Round and create DataFrame
df = pd.DataFrame({
    'Month': months,
    'Sales': sales,
    'Profit': profit.round(2)
})

df.to_csv("sample_sales_data.csv", index=False)
df.head()
print("Sample sales data generated and saved to 'sample_sales_data.csv'.")
print(df.head())
print("DataFrame shape:", df.shape)
```

```python
import matplotlib.pyplot as plt 

plt.style.use('ggplot')
df = pd.read_csv("sample_sales_data.csv")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Month'],df['Sales'], marker='o', label='Sales', color='royalblue')
plt.plot(df['Month'],df['Profit'], marker='s', label='profit', color='seagreen')


# Highlight best/worst month
best_month = df['Sales'].idxmax()
worst_month = df['Sales'].idxmin()  

plt.scatter(df.loc[best_month,'Month'], df.loc[best_month,'Sales'],color='gold', s=150, label='Best Month')

plt.scatter(df.loc[worst_month,'Month'], df.loc[worst_month,'Sales'],color='red', s=150, label='Worst Month')

plt.title('Monthly Sales and Profit Analysis', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Amount (₹)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df['Sales'], labels=df['Month'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Sales Share per Month", fontsize=16)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Sales'], df['Profit'], color='mediumpurple', s=100)
plt.title("Sales vs Profit", fontsize=16)
plt.xlabel("Sales (₹)")
plt.ylabel("Profit (₹)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

# Histogram
plt.subplot(1,2,1)
plt.hist(df['Profit'], bins=8, color='orange', edgecolor='black')
plt.title("Profit Distribution")

# Box plot
plt.subplot(1,2,2)
plt.boxplot(df['Profit'], patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title("Profit Spread")

plt.tight_layout()
plt.show()
```
